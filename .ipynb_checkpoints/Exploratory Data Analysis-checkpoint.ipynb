# What our research question is? 

State-of-the-art researches questioning fairness of COMPAS scores mainly focus on racial bias. We are interested in how other factors, such as age, gender, marital status, and sentence type, might also result in biased COMPAS scores.

COMPAS is a black box. We do not really know how the scoring system works: how many features it is using, what exactly are the features, etc.

# How we collected our data?

COMPAS itself did not have an open database, but it does have a [technical document]( http://www.northpointeinc.com/files/technical_documents/FieldGuide2_081412.pdf) for how to intepret the scores. 

As for COMPAS scores and crime data, we are drawing from a [open sourced project on github done by ProPublica](https://github.com/propublica/compas-analysis). We believe comparing to this, there is no better alternative. Since COMPAS is a black box and is a commercial software, it is challenging for us to request data directly from its developer. We also looked into other researches on COMPAS, but they gave way to ProPublica either because of smaller dataset or less complete inmate profiles.

The ProPublica dataset consists mainly of three files: `compas-scores-raw.csv`, `compas-scores-two-years.csv`, and `compas-scores-two-years-violent.csv`. 

Although we do not data ourselves, instead we use data published at [Propublica's github](https://github.com/propublica/compas-analysis), we still try to make sure that the data we use were ethically collected.

## How ProPublica [collected its data](https://www.propublica.org/article/how-we-analyzed-the-compas-recidivism-algorithm)?

ProPublica's data are from [Broward County](https://en.wikipedia.org/wiki/Broward_County,_Florida) in Florida, which  primarily uses COMPAS scores to determine whether to release or detain a defendant before their trial. ProPublica acquired COMPAS scores by an open request to the developer of COMPAS. As for profiles of inmates, ProPublica accessed through [Broward County Clerkâ€™s Office](https://www.clerk-17th-flcourts.org/Web2/CaseSearch/Results)

By looking into that, we become more confident that our research will not infringe upon citizens' privacy.

#### Features

"Agency_Text" describes the context where people was assessed by COMPAS. 

"LastName" and "FirstName" are name infos. 

"Sex_Code_Text" gives gender infos.

"Ethnic_Code_Text" gives racial infos. 

"DateOfBirth" gives age infos. 

"AssessmentReason" are all "intake".

"Language" are either English or Spanish.

"MaritalStatus" describes marital status. 

"DecileScore" are in a scale of 1 to 10 and "ScoreText" gives a category based on DecileScore. 

While past researches questioning fairness of COMPAS scores focus on racial factor, we want to investigate into other factors such as sex, age, language, marital status, past_crimes, etc.

# Applying `Aequitas` for some initial results

```python
compas_two_year.head()
```

__We selected sex, age, and race as the attributes in question:__

```python
df_eq = compas_two_year.loc[:, ['two_year_recid', 'decile_score', 'sex', 'age_cat', 'race']]
df_eq = df_eq.rename({'two_year_recid': 'label_value', 'decile_score': 'score'}, axis='columns')
df_eq.head()
```

```python
df_eq.to_csv('df_eq.csv')
```

We ran `Aequitas` from command line and obtained the result.

```python
df_eq_aequitas = pd.read_csv('df_eq_aequitas.csv')
```

__Because the actions to take is punitive, we are interested in false positive rate parity.__

We select `male` as the reference. The acceptable 80% to 120% interval is marked by the red lines:

```python
sns.barplot(x='fpr', y='attribute_value', data=df_eq_aequitas.loc[4:5, ['attribute_value', 'fpr']])
plt.axvline(df_eq_aequitas.loc[5, 'fpr'] * 0.8, c='r')
plt.axvline(df_eq_aequitas.loc[5, 'fpr'] * 1.2, c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('Sex');
```

For age, we select `25-45` as the reference:

```python
sns.barplot(x='fpr', y='attribute_value', data=df_eq_aequitas.loc[6:8, ['attribute_value', 'fpr']])
plt.axvline(df_eq_aequitas.loc[6, 'fpr'] * 0.8, c='r')
plt.axvline(df_eq_aequitas.loc[6, 'fpr'] * 1.2, c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('Age');
```

We select `Caucasian` for reference for race:

```python
sns.barplot(x='fpr', y='attribute_value', data=df_eq_aequitas.loc[9:, ['attribute_value', 'fpr']])
plt.axvline(df_eq_aequitas.loc[11, 'fpr'] * 0.8, c='r')
plt.axvline(df_eq_aequitas.loc[11, 'fpr'] * 1.2, c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('Race');
```

__Sadly, none of them pass the audit.__
